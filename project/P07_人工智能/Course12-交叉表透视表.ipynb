```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./data/stock_day.csv")
```

```python
data
```

```python
# 三元运算符，添加股票是否涨跌，涨为1，跌为0，生成新的列，列索引设置为"posi_neg"
data['posi_neg'] = np.where(data['p_change']>0,1,0)
```

```python
data.head()
```

```python
# 生成新的列，从索引的日期数据中转换出星期，生成新的列
data['week'] = pd.to_datetime(data.index).weekday
```

```python
data
```

```python
count = pd.crosstab(data['week'],data['posi_neg'])
```

```python
count
```

```python
# 先运算在求和
sum1 = count.sum(axis=1).astype(np.float32)
```

```python
sum1
```

```python
pro = count.div(sum1,axis=1)
```

```python
pro = pro.dropna(axis=1)
```

```python
# 生成柱状图，kind属性设置图形的类别，stacked属性设置图像是否叠加显示
# 默认为False不叠加显示
pro.plot(kind='bar',stacked=True,figsize=(20,10),fontsize=20)
```
