```python
from sklearn.feature_extraction import DictVectorizer

data_dict=[{"city":"","temperature":100},
     {"city":"","temperature":60},
     {"city":"","temperature":30}]
def dictvec():
    """
    
    :return None
    """
    # 
    dict = DictVectorizer(sparse=False)
    # sparseFalse
    # fit_transform
    data=dict.fit_transform(data_dict)
    print(dict.get_feature_names())
    print(dict.inverse_transform(data))
    print(data)
    return None
if __name__ == "__main__":
    dictvec()
```

```python
# 
from sklearn.feature_extraction.text import CountVectorizer

data_eg = ["life is short,I like python",
          "life is too long,I dislike python"]
def countvec():
    """
    
    :return None
    """
    text=CountVectorizer()
    data=text.fit_transform(data_eg)
    print(text.get_feature_names())
    # 
    # 
    # # jieba
    print(text.inverse_transform(data))
    print(data.toarray())
    return None
if __name__ == "__main__":
    countvec()
```
