```python
def wind_spddir_to_uv(wspd,wdir):
    """
    calculated the u and v wind components from wind speed and direction
    Input:
        wspd: wind speed
        wdir: wind direction
    Output:
        u: u wind component
        v: v wind component
    """

    rad = 4.0*np.arctan(1)/180.
    u = -wspd*np.sin(rad*wdir)
    v = -wspd*np.cos(rad*wdir)

    return u,v

# 
def _vor(udata, vdata):
    du_dy = np.gradient(udata, 1000, axis=0)
    dv_dx = np.gradient(vdata, 1000, axis=1)
    Vol = (dv_dx - du_dy)
    return Vol


def _div(udata, vdata):
    du_dx = np.gradient(udata, 1000, axis=1)
    dv_dy = np.gradient(vdata, 1000, axis=0)
    DIV = du_dx - dv_dy
    return DIV
```
